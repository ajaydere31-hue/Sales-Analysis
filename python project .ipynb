```python
numeric_cols = df.select_dtypes(include='number').columns
```

```python
numeric_cols
```

```python
state_pivot = pd.pivot_table(df,index='State',values=numeric_cols,aggfunc=['sum','mean'])
```

```python
state_pivot
```

```python
labels = state_pivot['mean']['Sales'].index.to_list()
```

```python
colors = sns.color_palette('pastel')[0:5]
```

```python
plt.pie(state_pivot['mean']['Sales'],labels=labels,colors=colors,autopct='%1.1f%%')
#plt.pie(state_pivot['mean']['Sales'],autopct='%1.1f%%')
plt.show()
```

```python
group_pivot = pd.pivot_table(df,index='Group',values=numeric_cols,aggfunc=['sum','mean'])
```
